```
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```
# Dataset
DS = "drive/MyDrive/Projetos/autos.csv"
df = pd.read_csv(DS,encoding = "latin-1")
```

```
df.head()
```

```
df.describe()
```

```
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
sns.set_style("white")
plt.figure(figsize=(12,8))
sns.barplot(x='fuelType', y='price', data=df, hue='gearbox')
plt.ylabel("Preço", fontsize=16)
plt.xlabel("Tipo do combustível", fontsize=16)
plt.title("Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio", fontsize=16)
plt.show()
```

```
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
plt.figure(figsize=(12,8))
sns.barplot(x='vehicleType', y='powerPS', data=df, hue='gearbox')
plt.xlabel('Tipo do veículo', fontsize=16)
plt.ylabel('Potência Média', fontsize=16)
plt.title('Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio', fontsize=16)
plt.show()
```
