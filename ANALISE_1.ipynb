```
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```
# Dataset
dataset = "drive/MyDrive/Projetos/autos.csv"
df = pd.read_csv(dataset,encoding = "latin-1")
```

```
df.describe()
```

```
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro

plt.figure(figsize=(16,12))
sns.set_style("white")
sns.distplot(df["yearOfRegistration"])
plt.xlabel("Ano de registro")
plt.ylabel("Densidade (KDE)")
plt.show()
```

```
# Crie um Boxplot para avaliar os outliers

plt.figure(figsize=(16,12))
sns.set_style("whitegrid")
sns.boxplot(x=df["vehicleType"], y=df["price"])
plt.xlabel("Tipo dos veículos", fontsize=22)
plt.ylabel("Preços dos veículos", fontsize=22)
plt.title("Comparação entre os preços dos veículos conforme o seu tipo", fontsize=22)
plt.show()

```

```
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria
sns.set_style("white")
fig = sns.catplot(x="vehicleType", data=df, kind="count", height=6, aspect=1.5)
fig.ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':16})
fig.ax.yaxis.set_label_text("Total de Veículos Para Venda", fontdict= {'size':16})
fig.ax.set_title("Contagem total de veículos à venda conforme o tipo de veículo",fontdict= {'size':18})
for p in fig.ax.patches:
    fig.ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+100))
```
