```
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```
# Dataset
dataset = "drive/MyDrive/Projetos/autos.csv"
df = pd.read_csv(dataset,encoding = "latin-1")
```

```
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.set_style("white")
fig = plt.figure(figsize=(12,8))
ax = sns.catplot(y='brand', data=df, kind='count', height=8, aspect=1.5)
ax.set_xlabels("Número de veículos")
ax.set_ylabels("Marca")
plt.show()

```

```
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

g=sns.factorplot(x='vehicleType', y=df['price'], hue='gearbox', data=df, size=6, kind="bar", height=8 , aspect = 1.5)
plt.xlabel("Tipo do veículo")
plt.ylabel("Preço Médio")
plt.title("Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio")
```
